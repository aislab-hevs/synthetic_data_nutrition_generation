```python
def form_probability_dict(proba_dict, widget_class, **kwargs):
    for k, v in proba_dict.items():
        if kwargs is not None:
            proba_dict[k] = widget_class(
                value=v,
                description=k.replace("_", " "), 
                disable=False,
                **kwargs
            )
        else:
            proba_dict[k] = widget_class(
                value=v,
                description=k.replace("_", " "), 
                disable=False
            )
    if len(proba_dict.keys()) > 3:
        return widgets.VBox(list(proba_dict.values()))
    return widgets.Box(list(proba_dict.values()))
    
```

```python
NUM_USERS = widgets.IntText(500, description="Total users")
display(NUM_USERS)
age_widgets = form_probability_dict(age_probabilities, widgets.FloatText)
gener_widgets = form_probability_dict(gender_probabilities, widgets.FloatText)
bmi_widgets = form_probability_dict(BMI_probabilities, widgets.FloatText)
allergies_widgets = form_probability_dict(allergies_probability_dict, widgets.FloatText)
food_restriction_widgets = form_probability_dict(food_restriction_probability_dict, widgets.FloatText)
meals_widget = form_probability_dict(meals_proba, widgets.FloatText)
flexi_accordion = widgets.Accordion(children=[form_probability_dict(flexi_probabilities[k], widgets.FloatText) for k in flexi_probabilities.keys()],
                                   titles= [k.replace("_", " ") for k in flexi_probabilities.keys()])
main_accordion = widgets.Accordion(children=[age_widgets, 
                                            gener_widgets, 
                                            bmi_widgets,
                                            allergies_widgets,
                                            food_restriction_widgets, 
                                            meals_widget, 
                                            flexi_accordion
                                           ], 
                                  titles=("Age", 
                                          "Gender", 
                                          "BMI",
                                          "Allergies",
                                          "Food restrictions",
                                          "Meal probability",
                                          "Flexible Probabilities"
                                         ))
display(main_accordion)
```

```python
def check_sum_probability(probability_dict):
    total_probability = sum([v.value for v in probability_dict.values()])
    if total_probability == 1.0:
        return True
    else:
        return False
    
```

```python
valid = widgets.Valid(
    value=check_sum_probability(age_probabilities),
    description='Valid!',
)
```
